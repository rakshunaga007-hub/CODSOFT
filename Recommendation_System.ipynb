```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = {
    'title': ['Avatar', 'Titanic', 'The Avengers', 'Iron Man', 'Thor', 'Hulk', 'Inception'],
    'genre': [
        'Action Adventure Sci-Fi',
        'Drama Romance',
        'Action Adventure Sci-Fi',
        'Action Sci-Fi',
        'Action Adventure Fantasy',
        'Action Sci-Fi',
        'Action Sci-Fi Thriller'
    ]
}

df = pd.DataFrame(data)

vectorizer= TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['genre'])

similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

def recommend_movie(movie_title, num_recommendations=3):
    if movie_title not in df['title'].values:
        return f"Movie '{movie_title}' not found in database."

    movie_index = df.index[df['title'] == movie_title][0]
    similarity_scores = list(enumerate(similarity_matrix[movie_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i, score in similarity_scores[1:num_recommendations + 1]:
        recommendations.append(df.iloc[i]['title'])

    return recommendations

movie = "Thor"
print(f"Recommended movies similar to '{movie}':")
print(recommend_movie(movie, 3))
```
